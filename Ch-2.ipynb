#  Data Visualization

Networkx

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,1)
G.add_edges_from([(1,4),(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(G,node_size=850,node_color="red",width=5,edge_color="blue")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,1)
G.add_edges_from([(1,4),(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(G,node_size=850,node_color="red",width=5,edge_color="blue")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([0,1,2])
l=[[1,0,0,1],[0,1,1,1],[1,1,1,0]]
for i in range(0,3):
    for j in range(0,3):
        if l[i][j]==1:
            G.add_edge(i,j)
nx.draw(G,with_labels=True)
plt.show()
```

Line Graph

```python
v1=[1,5,8,9]
v2=[3,8,9,2]

plt.plot(range(1,5),v1)
plt.plot(range(1,5),v2)
plt.savefig("fig1.png")
```

```python
v=[0,5,8,9,2,0,3,10]
ax=plt.axes()
ax.set_xticks([1,2,3,4,5,6,7,8])
ax.set_yticks([0,1,2,3,4,5,6,7])
plt.plot(range(1,9),v,"o--r")
plt.show()
```

```python
v=[0,5,8,9,2,0,3,10]
plt.annotate(xy=[1,1],text="First Entry")
plt.plot(range(1,9),v)
plt.legend(["First"],loc=0)
plt.show()
```
