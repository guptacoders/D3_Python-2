```python
df2=pd.read_csv("auto-mpg.csv")
print(df2)
```

```python
df2[df2['horsepower']=='?']
```

```python
df2['horsepower']=pd.to_numeric(df2['horsepower'].replace("?",0.0))
```

```python
print(df2.to_string())
```

```python
print(df2.info())
```

```python
quart1=df2['acceleration'].quantile(0.25)
quart3=df2['acceleration'].quantile(0.75)
IQR=quart3-quart1
low_val=quart1-1.5*IQR
high_val=quart3+1.5*IQR
df2=df2.loc[(df2['acceleration']<low_val)|(df2['acceleration']>high_val)]
print(df2)
```

```python
quart1=df2['acceleration'].quantile(0.25)
quart3=df2['acceleration'].quantile(0.75)
IQR=quart3-quart1
low_val=quart1-1.5*IQR
high_val=quart3+1.5*IQR
df2=df2.loc[(df2['acceleration']>=low_val)&(df2['acceleration']<=high_val)]
print(df2)
```
