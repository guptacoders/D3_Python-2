```python
import pandas as pd

df=pd.read_csv("diabetes_unclean - Copy.csv")
print(df.to_string())
```

```python
# Display the specific statistiic that are relevant for numeric type column
df.describe()
```

```python
# How many rows and column givenin dataset

df.shape
```

```python
# Ques: Check the missing value of dataset
df.isna().sum()
```

```python
# Ques: Replace the missing value in collumn "HbA1c" with their mean value

df['HbA1c']=df['HbA1c'].fillna(df['HbA1c'].mean())
```

```python
# Ques: Drop the missing values of other columns

df=df.dropna()
```

```python
# Ques: Check the informatiomn of your dataset
df.info()
```

```python
# Ques: In class column "N " and "Y" it should be replaced with "N" and "Y


df["CLASS"]=df["CLASS"].replace("N ","N")
df["CLASS"]=df["CLASS"].replace("Y ","Y")
```

```python
# Checking that it was changed or not
df.CLASS.unique()
```

```python
# Ques: Show the correlation between variables

df.corr()
# For new versions: df.corr(numeric_only=True)
```

```python
# Draw the box plot for columns AGE,Urea ,HbA1C,Chol,TG,HDL,LDL,VLDL,BMI

import matplotlib.pyplot as plt
import pandas as pd

col = ['AGE', 'Urea', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
df.boxplot(col)
plt.show()
```

```python
# Visualize the columns Urea ,HbA1C,TG,BMI for different AGE using parallel coordinates

import matplotlib.pyplot as plt
import pandas as pd
pd.plotting.parallel_coordinates(df,'AGE',['Urea','HbA1c','TG','BMI'])
plt.show()
```

```python
# Remove the rows whose gender column has 'f' and give frequency of 'F' and 'M' in different class

df=df.drop(df[df['Gender']=='f'].index)
pd.crosstab(df['Gender'],df["CLASS"])
```

```python
# Remove the Outliers of HbA1c and give shape of df

quart1 = df['HbA1c'].quantile(0.25)
quart3 = df['HbA1c'].quantile(0.75)
IQR = quart3 - quart1
low_val = quart1 - 1.5 * IQR
high_val = quart3 + 1.5 * IQR

df = df.loc[(df['HbA1c'] >= low_val) & (df['HbA1c'] <= high_val)]
print(df)

```
